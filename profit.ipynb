```python
import yfinance as yf
import pandas as pd
from pandas_datareader import data as pdr
import datetime as dt
import numpy as np
```

```python
endDate = dt.datetime.now() + dt.timedelta(hours = 9) #  

startDate = endDate - dt.timedelta(days = 365*5)
print(endDate, startDate)

stocks = ['SPY', 'SCHD', 'QQQ']

df = yf.download(stocks, start = startDate, end = endDate)
# print(df.head())

adj_close_prices = df['Close']
adj_close_prices.head()

log_returns = np.log(adj_close_prices/adj_close_prices.shift(1))
log_returns.head()

cum_log_returns = log_returns.cumsum()

cum_log_returns.plot(title = "cumulative returns", figsize = (10, 6))
```
